```python
def replace_columns(df, new_columns): # el primer argumento tiene que ser una DF y el segundo un diccionario
    new_col_names = {original_name : category + '_' + new_name 
                     for category, cols in new_columns.items() 
                     for original_name, new_name in cols.items()}
    return df.rename(columns = new_col_names) #new_col_names es un diccionario tal que {old_name:new:name}.
                                                # es el diccionario que definimos en la primera linea de esta funcion



```

```python
df_renamed = replace_columns(df, new_columns)
df_renamed.head()
```

```python
# que tamano tiene el DF?
print(f'{df_renamed.shape[0]} filas y {df_renamed.shape[1]} columnas')
# confirmamos que con los cambios que hicimos no modificamos el numero de filas y columnas de la DF
```
